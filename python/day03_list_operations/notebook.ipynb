# Day 3 — Python: Array & List Operations
**Topics:** sorting, slicing, list comprehensions, deduplication, stack pattern, two-pointer

## 1. List Basics & Mutations

```python
nums = [3, 1, 4, 1, 5, 9, 2, 6]
print('original:', nums)

nums.append(7)
print('after append(7):', nums)

nums.insert(0, 0)
print('after insert(0, 0):', nums)

popped = nums.pop()
print('after pop():', nums, '  popped:', popped)

nums.remove(1)   # removes first occurrence
print('after remove(1):', nums)
```

## 2. sorted() vs list.sort()

```python
salaries = [95000, 72000, 85000, 95000, 60000, 85000]

# sorted() — new list, original untouched
desc = sorted(salaries, reverse=True)
print('sorted() desc:', desc)
print('original still:', salaries)

# list.sort() — in-place, returns None
copy = salaries[:]
result = copy.sort(reverse=True)
print('sort() returned:', result)   # None!
print('copy after sort():', copy)
```

## 3. Slicing

```python
nums = [10, 20, 30, 40, 50, 60, 70]

print(nums[1:4])    # [20, 30, 40]
print(nums[:3])     # [10, 20, 30]
print(nums[-3:])    # [50, 60, 70]
print(nums[::2])    # every 2nd
print(nums[::-1])   # reversed

copy = nums[:]
copy[0] = 999
print('copy modified:', copy)
print('original unchanged:', nums)
```

## 4. List Comprehensions

```python
# Squares of even numbers
even_sq = [x**2 for x in range(1, 11) if x % 2 == 0]
print('even squares:', even_sq)

# Clean a messy list
raw = ['  Alice ', ' BOB  ', 'carol  ', None, '  DAVE']
clean = [s.strip().title() for s in raw if s is not None]
print('cleaned:', clean)

# Flatten nested list
batches = [[1, 2, 3], [4, 5], [6, 7, 8, 9]]
flat = [x for batch in batches for x in batch]
print('flattened:', flat)
```

## 5. Deduplication — set vs dict.fromkeys

```python
ids = [101, 203, 101, 305, 203, 101, 407]

# set — does NOT preserve order
print('set result:', list(set(ids)))          # order unpredictable

# dict.fromkeys — preserves insertion order (Python 3.7+)
print('dict.fromkeys:', list(dict.fromkeys(ids)))   # [101, 203, 305, 407]
```

## 6. Second Highest — sort once

```python
salaries = [95000, 72000, 85000, 95000, 60000, 85000]

# One sort on unique values
unique = sorted(set(salaries), reverse=True)
print('unique sorted:', unique)
print('second highest:', unique[1])
```

## 7. Stack Pattern — list as stack

```python
stack = []

stack.append('a')   # push
stack.append('b')
stack.append('c')
print('stack:', stack)

top = stack.pop()   # pop from end — O(1)
print('popped:', top, '  stack:', stack)
print('peek (stack[-1]):', stack[-1])
```

## 8. Stack Problem — Balanced Brackets

```python
def is_balanced(s):
    stack = []
    pairs = {')': '(', ']': '[', '}': '{'}
    for ch in s:
        if ch in '([{':
            stack.append(ch)
        elif ch in ')]}' :
            if not stack or stack[-1] != pairs[ch]:
                return False
            stack.pop()
    return len(stack) == 0

tests = ['({[]})', '({[})', '((()))', '(((']
for t in tests:
    print(f'{t!r:15} -> {is_balanced(t)}')
```

## 9. Monotonic Stack — Daily Temperatures

```python
def days_until_warmer(temps):
    result = [0] * len(temps)
    stack  = []    # stack of indices waiting for a warmer day

    for i, temp in enumerate(temps):
        # today is warmer than the day at stack top
        while stack and temps[stack[-1]] < temp:
            prev = stack.pop()
            result[prev] = i - prev
        stack.append(i)

    return result

temps = [73, 74, 75, 71, 69, 72, 76, 73]
result = days_until_warmer(temps)
print('temps: ', temps)
print('result:', result)
# expected: [1, 1, 4, 2, 1, 1, 0, 0]
```

## 10. Two-Pointer Pattern

```python
def two_sum_sorted(nums, target):
    left, right = 0, len(nums) - 1
    while left < right:
        s = nums[left] + nums[right]
        if s == target:
            return (nums[left], nums[right])
        elif s < target:
            left += 1
        else:
            right -= 1
    return None

nums = sorted([3, 7, 1, 8, 2, 5])
print('sorted input:', nums)
print('pair summing to 9:', two_sum_sorted(nums, 9))
```
