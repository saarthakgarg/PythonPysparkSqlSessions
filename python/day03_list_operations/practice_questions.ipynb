````python
# Warm-up 1: dict.fromkeys preserves insertion order (Python 3.7+)
ids = [101, 203, 101, 305, 203, 407]
d   = dict.fromkeys(ids)
print('dict from ids:', d)          # keys in insertion order, all values None
print('list of keys:', list(d))
````

````python
# Warm-up 2: set loses order — don't use for order-preserving dedup
ids = [5, 3, 1, 3, 5, 7, 1]
print('set (order lost):', list(set(ids)))           # could be anything
print('dict.fromkeys (order kept):', list(dict.fromkeys(ids)))  # [5, 3, 1, 7]
````

````python
# Warm-up 3: manual seen-set approach (shows the underlying logic)
ids  = [101, 203, 101, 305, 203, 407]
# seen = set()
out  = []
for x in ids:
    if x not in out:
        # seen.add(x)
        out.append(x)
print('manual dedup:', out)

# initial value = seen = (), out=[]
# after 1st iteration-> seen = (101), out=[101]
# after 2nd iteration -> seen= (101, 203), out=[101, 201]
# after 3rd iteration -> seen = (101, 203, 305), output = [101, 201, 305]
````

# Question 3 🟡 — Days Until Warmer Temperature (Monotonic Stack)

## Concept Warm-up

````python
# Warm-up 1: list as stack — push and pop
stack = []
for i in range(5):
    stack.append(i)
    print(f'push {i} → stack: {stack}')

while stack:
    top = stack.pop()
    print(f'pop {top} ← stack: {stack}')
````

````python
# Warm-up 2: stack of indices — peek at index and look up original array
temps  = [73, 74, 75, 71, 69]
stack  = []
result = [0] * len(temps) # [0, 0, 0, 0, 0]

for i, t in enumerate(temps):
    print(f'Day {i}: temp={t}  stack(indices)={stack}')
    # peek: what's the top of the stack?
    if stack:
        prev_idx  = stack[-1]
        prev_temp = temps[prev_idx]
        print(f'  top of stack → Day {prev_idx}, temp {prev_temp}')
        if t > prev_temp:
            print(f'  Today is warmer! Day {prev_idx} waited {i - prev_idx} days')
    stack.append(i)
````

````python
# Warm-up 3: the while loop pops ALL days colder than today
# Example: temps = [69, 71, 73] — every previous day is colder than today
temps  = [69, 71, 73]
stack  = []
result = [0] * len(temps) # [0, 0, 0]

for i, t in enumerate(temps):
    print(f'Day {i}: temp={t}')
    while stack and temps[stack[-1]] < t:
        prev = stack.pop()
        result[prev] = i - prev
        print(f'  Resolved Day {prev}: waited {result[prev]} days')
    stack.append(i)

print('result:', result)   # [2, 1, 0]

stack = [0]
````
