## 1. Dict Basics — Create, Read, Write, Delete

```python
# Create
emp = {'name': 'Alice', 'dept': 'Engineering', 'salary': 95000}
print('dict:', emp)

# Read
print(emp['name'])            # 'Alice'
print(emp.get('age', 0))      # 0  — safe, no KeyError

# Write
emp['salary'] = 100000        # update
emp['city']   = 'New York'    # new key
print('after update:', emp)

# Delete
emp.pop('city')
print('after pop:', emp)

# Membership
print('name in emp:', 'name' in emp)
print('age  in emp:', 'age'  in emp)
```

## 2. Iterating — keys, values, items

```python
scores = {'Alice': 95, 'Bob': 72, 'Carol': 88}

print('--- keys ---')
for name in scores:          # default iterates keys
    print(name)

print('--- items ---')
for name, score in scores.items():
    print(f'{name}: {score}')

print('--- values ---')
for score in scores.values():
    print(score)
```

## 3. Frequency Map (Counter Pattern)

```python
sentence = "the cat sat on the mat the" 

# Manual frequency dict
freq = {}
for word in sentence.split():  # using split ['the', 'cat', 'sat', 'on', 'the', 'mat', 'the']
    freq[word] = freq.get(word, 0) + 1    # get returns 0 if missing  1st iteration = {'the': 3, 'cat': 1, 'sat': 1, 'on': 1, 'mat': 1, }

print('freq:', freq)
print('most common:', max(freq, key=freq.get))

# Same using collections.Counter
from collections import Counter
c = Counter(sentence.split())
print('Counter:', c)
print('most_common(1):', c.most_common(1))
```

## 4. HashMap Pattern — Two Sum O(n)

```python
# Brute force O(n²) — for comparison
def two_sum_brute(nums, target):
    for i in range(len(nums)):
        for j in range(i + 1, len(nums)):
            if nums[i] + nums[j] == target:
                return [i, j]
    return []

print('brute force:', two_sum_brute([2, 7, 11, 15], 9))
```

```python
# HashMap O(n) — for each num, check if complement already seen
def two_sum(nums, target):
    seen = {}             # {value: index}
    for i, num in enumerate(nums):
        complement = target - num
        if complement in seen:
            return [seen[complement], i]
        seen[num] = i
    return []

print(two_sum([2, 7, 11, 15], 9))    # [0, 1]
print(two_sum([3, 2, 4], 6))         # [1, 2]
print(two_sum([3, 3], 6))            # [0, 1]

# Trace the logic
nums, target = [2, 7, 11, 15], 9
seen = {}
for i, num in enumerate(nums):
    comp = target - num
    print(f'i={i}  num={num}  need={comp}  seen={seen}  found={comp in seen}')
    if comp in seen:
        print(f'  -> pair: [{seen[comp]}, {i}]')
        break
    seen[num] = i
```

## 5. setdefault — GroupBy Pattern

```python
employees = [
    {'name': 'Alice',   'dept': 'Engineering', 'salary': 95000},
    {'name': 'Bob',     'dept': 'Marketing',   'salary': 72000},
    {'name': 'Carol',   'dept': 'Engineering', 'salary': 88000},
    {'name': 'David',   'dept': 'HR',          'salary': 65000},
    {'name': 'Eve',     'dept': 'Marketing',   'salary': 78000},
    {'name': 'Frank',   'dept': 'Engineering', 'salary': 101000},
]

# Group names by department
by_dept = {}
for emp in employees:
    by_dept.setdefault(emp['dept'], []).append(emp['name'])

for dept, names in sorted(by_dept.items()):
    print(f'{dept}: {names}')
```

```python
# Accumulate headcount + total salary, then compute avg
dept_stats = {}
for emp in employees:
    dept = emp['dept']
    dept_stats.setdefault(dept, {'headcount': 0, 'total_salary': 0.0})
    dept_stats[dept]['headcount']    += 1
    dept_stats[dept]['total_salary'] += emp['salary']

for dept, s in sorted(dept_stats.items()):
    avg = round(s['total_salary'] / s['headcount'], 2)
    print(f'{dept:<15} headcount={s["headcount"]}  avg_salary={avg}')
```

## 6. defaultdict — Cleaner GroupBy

```python
from collections import defaultdict

# defaultdict(list) — missing keys auto-init to []
by_dept = defaultdict(list)
for emp in employees:
    by_dept[emp['dept']].append(emp['name'])

for dept, names in sorted(by_dept.items()):
    print(f'{dept}: {names}')
```

## 7. Dictionary Comprehension

```python
# name → salary for all employees
salary_map = {emp['name']: emp['salary'] for emp in employees}
print('salary_map:', salary_map)

# Filter: keep only Engineering
eng_salaries = {emp['name']: emp['salary'] for emp in employees if emp['dept'] == 'Engineering'}
print('engineering:', eng_salaries)

# Invert a dict
original = {'a': 1, 'b': 2, 'c': 3}
inverted = {v: k for k, v in original.items()}
print('inverted:', inverted)
```

## 8. DE Use Case — Word Frequency in a Log Field

```python
# Simulate log error messages
logs = [
    'ERROR connection timeout',
    'WARN disk usage high',
    'ERROR authentication failed',
    'INFO pipeline started',
    'ERROR connection timeout',
    'ERROR null pointer',
    'WARN disk usage high',
    'ERROR connection timeout',
]

# Count log level frequency
level_count = {}
for log in logs:
    level = log.split()[0]
    level_count[level] = level_count.get(level, 0) + 1

print('Log levels:', level_count)

# Count unique error messages
error_msgs = Counter(
    log[len('ERROR '):] for log in logs if log.startswith('ERROR')
)
print('\nError message frequency:')
for msg, count in error_msgs.most_common():
    print(f'  {count}x  {msg}')
```
