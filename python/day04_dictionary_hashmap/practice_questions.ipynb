# Question 1 🟢 — Two Sum Using HashMap

## Concept Warm-up

````python
# Warm-up 1: dict.get(key, default) — safe read without KeyError
d = {'apple': 3, 'banana': 5}

print(d.get('apple', 0))    # 3   — key exists
print(d.get('mango', 0))    # 0   — key missing, returns default
print(d.get('mango'))       # None — key missing, no default
````

````python
# Warm-up 2: check membership — O(1) in dict vs O(n) in list
seen = {2: 0, 5: 1, 3: 2}   # {value: index}

print(3 in seen)     # True   — O(1)
print(9 in seen)     # False

# This is the core of Two Sum:
# if complement in seen: return [seen[complement], current_index]
````

````python
# Warm-up 3: enumerate gives (index, value) — both needed for Two Sum
nums = [2, 7, 11, 15]

for i, num in enumerate(nums):
    print(f'index={i}  value={num}  complement_for_target_9={9 - num}')
````

## Problem

Given a list of integers `nums` and an integer `target`, return the **two indices** whose values add up to `target`.
- Use a **HashMap** — do NOT use a nested loop  
- Return `[]` if no pair exists

```
two_sum([2, 7, 11, 15], 9)  →  [0, 1]   # 2 + 7 = 9
two_sum([3, 2, 4], 6)       →  [1, 2]   # 2 + 4 = 6
two_sum([3, 3], 6)          →  [0, 1]
```

**Hint:** For each number, the number you need is `target - num`. Store `{value: index}` as you iterate and check if the complement is already in the dict.

````python
def two_sum(nums: list, target: int) -> list:
    # YOUR CODE HERE
    seen = {} # hashmap
    for i, num in enumerate(nums):
        
        complement = target-num
        if complement in seen:
            return [seen[complement], i]
        seen[num]=i
    return []

    
# two_sum([2, 7, 11, 15], 9)

# print(two_sum([3, 2, 4], 6) )

# Tests
assert two_sum([2, 7, 11, 15], 9)  == [0, 1]
assert two_sum([3, 2, 4], 6)       == [1, 2]
assert two_sum([3, 3], 6)          == [0, 1]
assert two_sum([1, 5, 3, 7], 10)   == [2, 3]
assert two_sum([1, 2, 3], 99)      == []
print('All tests passed!')
````

# Question 2 🟢 — Highest Frequency Word

## Concept Warm-up

````python
# Warm-up 1: build a frequency dict with .get(key, 0)
words = ['the', 'cat', 'sat', 'on', 'the', 'mat', 'the']

freq = {}
for word in words:
    freq[word] = freq.get(word, 0) + 1

print('freq:', freq)
# 'the' → 3, 'cat' → 1, 'sat' → 1, 'on' → 1, 'mat' → 1
````

````python
# Warm-up 2: max(dict, key=dict.get) returns the KEY with the max VALUE
freq = {'the': 3, 'cat': 1, 'sat': 1, 'on': 1, 'mat': 1}

print('max value:', max(freq.values()))       # 3
print('key with max value:', max(freq, key=freq.get))  # 'the'
````

````python
# Warm-up 3: str.lower().split() handles case-insensitivity
sentence = "The cat sat on the mat THE"
words = sentence.lower().split()
print('words:', words)
# ['the', 'cat', 'sat', 'on', 'the', 'mat', 'the']
````

# Question 3 🟡 — Group Employees by Department

## Concept Warm-up

````python
# Warm-up 1: setdefault — init missing key and return value in one call
d = {}

d.setdefault('Engineering', {'count': 0, 'total': 0.0})
d['Engineering']['count'] += 1
d['Engineering']['total'] += 95000

d.setdefault('Engineering', {'count': 0, 'total': 0.0})  # already exists — no-op
d['Engineering']['count'] += 1
d['Engineering']['total'] += 88000

print(d)
````

````python
# Warm-up 2: iterate a list of dicts and access fields
employees = [
    {'name': 'Alice', 'dept': 'Engineering', 'salary': 95000},
    {'name': 'Bob',   'dept': 'Marketing',   'salary': 72000},
]

for emp in employees:
    print(f"name={emp['name']}  dept={emp['dept']}  salary={emp['salary']}")
````
