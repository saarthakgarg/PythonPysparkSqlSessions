## 1. Creating Sets

```python
# From a list — deduplicates automatically
product_ids = ['P001', 'P002', 'P001', 'P003', 'P002']
unique_ids = set(product_ids)
print('Unique IDs:', unique_ids)          # order is not guaranteed
print('Count:', len(unique_ids))

# Empty set — NEVER use {} for empty set (that creates a dict)
empty_set = set()
empty_dict = {}
print('\nTypes:', type(empty_set), type(empty_dict))
```

```python
# Add and remove
s = {1, 2, 3}

s.add(4)
print('After add(4):', s)

s.discard(99)          # safe — no error even though 99 is not in set
print('After discard(99):', s)

try:
    s.remove(99)       # KeyError — 99 not in set
except KeyError as e:
    print(f'remove() raised KeyError: {e}')
```

```python
# Membership — O(1) for set, O(n) for list
valid_statuses_set  = {'completed', 'pending', 'cancelled'}
valid_statuses_list = ['completed', 'pending', 'cancelled']

print('in set: ',  'completed' in valid_statuses_set)   # O(1)
print('in list:', 'completed' in valid_statuses_list)   # O(n)
print('not in set:', 'unknown' not in valid_statuses_set)
```

## 2. The Four Set Operations

```python
a = {1, 2, 3, 4}
b = {3, 4, 5, 6}

print('Union        (|):', a | b)              # {1, 2, 3, 4, 5, 6} — all from both
print('Intersection (&):', a & b)              # {3, 4}             — in BOTH
print('Difference   (-):', a - b)              # {1, 2}             — in A, not B
print('Sym diff     (^):', a ^ b)              # {1, 2, 5, 6}       — in one but not both

# Method syntax (same result)
print('\nMethod syntax:')
print('a.union(b):          ', a.union(b))
print('a.intersection(b):   ', a.intersection(b))
print('a.difference(b):     ', a.difference(b))
print('a.symmetric_diff(b):', a.symmetric_difference(b))
```

## 3. Deduplication — With and Without Order Preservation

```python
customer_ids = ['C003', 'C001', 'C002', 'C001', 'C003', 'C004']

# Fast deduplicate — loses original order
deduped_unordered = list(set(customer_ids))
print('Unordered dedup:', deduped_unordered)

# Preserve insertion order
seen = set()
deduped_ordered = []
for cid in customer_ids:
    if cid not in seen:
        seen.add(cid)
        deduped_ordered.append(cid)

print('Ordered dedup:  ', deduped_ordered)
```

## 4. Data Engineering Patterns — New / Deleted / Changed IDs

```python
previous_ids = {'C001', 'C002', 'C003'}
current_ids  = {'C002', 'C003', 'C004', 'C005'}

new_ids     = current_ids - previous_ids      # in current but not previous
deleted_ids = previous_ids - current_ids      # in previous but not current
retained    = previous_ids & current_ids      # in both
all_ids     = previous_ids | current_ids      # all unique IDs ever

print('New (to load):', sorted(new_ids))
print('Deleted:      ', sorted(deleted_ids))
print('Retained:     ', sorted(retained))
print('All:          ', sorted(all_ids))
```

## 5. First Non-Repeating Character Pattern

```python
from collections import Counter

def first_non_repeating(s):
    counts = Counter(s)
    for ch in s:
        if counts[ch] == 1:
            return ch
    return None

print(first_non_repeating('aabbcde'))   # 'c'
print(first_non_repeating('aabb'))      # None
print(first_non_repeating('stress'))    # 't'
```

```python
# Alternative: two-pass using seen_once and seen_more sets (no Counter import needed)
def first_non_repeating_v2(s):
    seen_once = set()
    seen_more = set()
    for ch in s:
        if ch in seen_more:
            continue
        if ch in seen_once:
            seen_once.discard(ch)
            seen_more.add(ch)
        else:
            seen_once.add(ch)
    for ch in s:
        if ch in seen_once:
            return ch
    return None

print(first_non_repeating_v2('aabbcde'))  # 'c'
print(first_non_repeating_v2('aabb'))     # None
```

## 6. Subset / Superset Checks

```python
required_cols = {'id', 'name', 'email'}
actual_cols   = {'id', 'name', 'email', 'city', 'joined_at'}

print('All required cols present:', required_cols.issubset(actual_cols))   # True
print('Actual is superset:       ', actual_cols.issuperset(required_cols)) # True

missing = required_cols - actual_cols
print('Missing cols:', missing)   # empty set → all present

# Practical: validate a pipeline output has all required columns
def validate_columns(required, actual):
    missing = set(required) - set(actual)
    if missing:
        raise ValueError(f'Missing columns: {missing}')
    print('All required columns present.')

validate_columns(required_cols, actual_cols)
```

## Quick Reference

| Operation | Syntax | Result |
|-----------|--------|--------|
| Union | `a \| b` | All elements from both |
| Intersection | `a & b` | Only in both |
| Difference | `a - b` | In A, not in B |
| Symmetric diff | `a ^ b` | In one but not both |
| Membership | `x in s` | O(1) |
| Add | `s.add(x)` | — |
| Safe remove | `s.discard(x)` | No error if missing |
| Strict remove | `s.remove(x)` | KeyError if missing |
