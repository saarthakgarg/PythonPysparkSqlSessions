````python
# --- assert ---
assert first_non_repeating('aabbcde') == 'c',   f"got {first_non_repeating('aabbcde')}"
assert first_non_repeating('stress')  == 't',   f"got {first_non_repeating('stress')}"
assert first_non_repeating('aabb')    is None,  f"got {first_non_repeating('aabb')}"
assert first_non_repeating('a')       == 'a',   f"got {first_non_repeating('a')}"
assert first_non_repeating('')        is None,  f"got {first_non_repeating('')}"
print('Q2 passed!')
````

### Solution

````python
# SOLUTION
def first_non_repeating(s):
    counts = Counter(s)         # one pass to count all
    for ch in s:                # second pass preserves original order
        if counts[ch] == 1:
            return ch
    return None

print(first_non_repeating('aabbcde'))   # 'c'
print(first_non_repeating('stress'))    # 't'
print(first_non_repeating('aabb'))      # None
````

````python
# SOLUTION
def customers_in_both(refunds, purchases):
    return set(refunds) & set(purchases)

def refund_only_customers(refunds, purchases):
    return set(refunds) - set(purchases)

refunds   = ['C001', 'C003', 'C005', 'C007']
purchases = ['C001', 'C002', 'C005', 'C008']

print('In both (flag for review):', customers_in_both(refunds, purchases))
print('Refund only (no purchase):', refund_only_customers(refunds, purchases))
````
