````python
# Warm-up 2: generator function with yield
def evens_up_to(n):
    for i in range(n):
        if i % 2 == 0:
            yield i

gen = evens_up_to(10)
print(next(gen))          # 0 — pulls one item
print(list(gen))          # [2, 4, 6, 8] — exhausts the rest
````
